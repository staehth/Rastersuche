```python
# Wortsuche in Raster

word = "ZHAW"

letters = [
    ['P', 'T', 'H', 'G', 'O', 'M', 'G', 'O'],
    ['F', 'U', 'I', 'L', 'N', 'E', 'D', 'N'],
    ['E', 'R', 'A', 'U', 'W', 'W', 'C', 'U'],
    ['O', 'T', 'I', 'M', 'A', 'A', 'R', 'T'],
    ['H', 'O', 'D', 'H', 'E', 'T', 'E', 'N'],
    ['N', 'K', 'Z', 'N', 'L', 'L', 'Y', 'O'],
    ['U', 'E', 'R', 'D', 'H', 'G', 'U', 'C'],
    ['S', 'T', 'E', 'A', 'A', 'E', 'M', 'F']
]
letters_found = []

wordLength, RowMax, ColMax = len(word), len(letters), len(letters[0])


def checkLetters(x,y,direction):

    Words = []
    Coords = []

    if direction == 'nord':
        end_r, end_c = x - (wordLength - 1), y
        if not (0 <= end_r < RowMax and 0 <= end_c < ColMax):
            return [],[]
        Word = ""
        Coord = []
        for z in range(wordLength):
            Word += (letters[x-z][y])
            Coord.append([x-z, y])
        Words.append(Word)
        Coords.append(Coord)

    if direction == 'ost':
        end_r = x
        end_c = y + wordLength - 1
        if not (0 <= end_r < RowMax and 0 <= end_c < ColMax):
            return [],[]
        Word = ""
        Coord = []
        for z in range(wordLength):
            Word += letters[x][y+z]
            Coord.append([x, y+z])
        Words.append(Word)
        Coords.append(Coord)

    if direction == 'sued':
        end_r = x + (wordLength - 1)
        end_c = y
        if not (0 <= end_r < RowMax and 0 <= end_c < ColMax):
            return [],[]
        Word = ""
        Coord = []
        for z in range(wordLength):
            Word += (letters[x+z][y])
            Coord.append([x+z, y])
        Words.append(Word)
        Coords.append(Coord)

    if direction == 'west':
        end_r = x
        end_c = y - (wordLength - 1)
        if not (0 <= end_r < RowMax and 0 <= end_c < ColMax):
            return [],[]
        Word = ""
        Coord = []
        for z in range(wordLength):
            Word += (letters[x][y-z])
            Coord.append([x, y-z])
        Words.append(Word)
        Coords.append(Coord)

    if direction == 'nordost':
        end_r = x - (wordLength - 1)
        end_c = y + (wordLength - 1)
        if not (0 <= end_r < RowMax and 0 <= end_c < ColMax):
            return [],[]
        Word = ""
        Coord = []
        for z in range(wordLength):
            Word += (letters[x-z][y+z])
            Coord.append([x-z, y+z])
        Words.append(Word)
        Coords.append(Coord)
   
    if direction == 'suedost':
        end_r = x + (wordLength - 1)
        end_c = y + (wordLength - 1)
        if not (0 <= end_r < RowMax and 0 <= end_c < ColMax):
            return [],[]
        Word = ""
        Coord = []
        for z in range(wordLength):
            Word += (letters[x+z][y+z])
            Coord.append([x+z, y+z])
        Words.append(Word)
        Coords.append(Coord)

    if direction == 'nordwest':
        end_r = x - (wordLength - 1)
        end_c = y - (wordLength - 1)
        if not (0 <= end_r < RowMax and 0 <= end_c < ColMax):
            return [],[]
        Word = ""
        Coord = []
        for z in range(wordLength):
            Word += (letters[x-z][y-z])
            Coord.append([x-z, y-z])
        Words.append(Word)
        Coords.append(Coord)

    if direction == 'suedwest':
        end_r = x + (wordLength - 1)
        end_c = y - (wordLength - 1)
        if not (0 <= end_r < RowMax and 0 <= end_c < ColMax):
            return [],[]
        Word = ""
        Coord = []
        for z in range(wordLength):
            Word += (letters[x+z][y-z])
            Coord.append([x+z, y-z])
        Words.append(Word)
        Coords.append(Coord)

    return Words, Coords

for richtung in ["nord","sued","west","ost","nordost","suedost","suedwest","nordwest"]:
    for x in range(RowMax):
        for y in range(ColMax):
            words,coords = checkLetters(x,y,richtung)

            for i in range(len(words)):
                if words[i] == word:
                    print(f"\"{words[i]}\" gefunden bei {x},{y} nach {richtung.upper()}: {coords[i]}")
                    print()
                    for row in range(RowMax):
                        for i,col in enumerate(range(ColMax)):
                            if [row,col] in coords[0]:
                                print(f" ({letters[row][col]}) ",end="")
                            else:
                                print(f"  {letters[row][col]}  ",end="")
                            if (i+1 == ColMax):
                                print()
                    print()

#def check_direction(x: int, y: int, direction: str, word_length: int, letters: list, row_max: int, col_max: int) -> tuple[list, list]:
#    """
#    Check for words in a specific direction from a starting position in a letter grid.
#    
#    Args:
#        x (int): Starting row position
#        y (int): Starting column position
#        direction (str): Direction to check ('nord', 'sued', 'west', 'ost', 'nordost', 'suedost', 'suedwest', 'nordwest')
#        word_length (int): Length of the word to check
#        letters (list): 2D grid of letters
#        row_max (int): Maximum number of rows in the grid
#        col_max (int): Maximum number of columns in the grid
#        
#    Returns:
#        tuple: A tuple containing (words_found, coordinates)
#            - words_found (list): List of words found in the specified direction
#            - coordinates (list): List of coordinate lists for each found word
#    """
#    # Define direction vectors for each possible direction
#    direction_vectors = {
#        'nord': (-1, 0),
#        'sued': (1, 0),
#        'west': (0, -1),
#        'ost': (0, 1),
#        'nordost': (-1, 1),
#        'suedost': (1, 1),
#        'nordwest': (-1, -1),
#        'suedwest': (1, -1)
#    }
#    
#    if direction not in direction_vectors:
#        return [], []
#    
#    # Get direction vector
#    dr, dc = direction_vectors[direction]
#    
#    # Calculate end position
#    end_r = x + dr * (word_length - 1)
#    end_c = y + dc * (word_length - 1)
#    
#    # Check if end position is within bounds
#    if not (0 <= end_r < row_max and 0 <= end_c < col_max):
#        return [], []
#    
#    # Extract word and coordinates
#    word = ""
#    coords = []
#    for z in range(word_length):
#        current_r = x + dr * z
#        current_c = y + dc * z
#        word += letters[current_r][current_c]
#        coords.append([current_r, current_c])
#    
#    return [word], [coords]
#
## Usage example:
#for direction in ["nord", "sued", "west", "ost", "nordost", "suedost", "suedwest", "nordwest"]:
#    for x in range(row_max):
#        for y in range(col_max):
#            words, coords = check_direction(x, y, direction, word_length, letters, row_max, col_max)
#            
#            for i in range(len(words)):
#                if words[i] == word:
#                    print(f"\"{words[i]}\" gefunden bei {x},{y} nach {direction.upper()}: {coords[i]}")
#                    print()
#                    for row in range(row_max):
#                        for i, col in enumerate(range(col_max)):
#                            if [row, col] in coords[0]:
#                                print(f" ({letters[row][col]}) ", end=" ")
#                            else:
#                                print(f"  {letters[row][col]}  ", end=" ")
#                            if (i+1 == col_max):
#                                print()
#                    print()
```
